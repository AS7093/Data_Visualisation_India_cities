# Area Plots<a id="6"></a>

In the last, we created a line plot that visualized the population of Kerala. With a little modification to the code, we can visualize this plot as a cumulative plot, also knows as a **Stacked Line Plot** or **Area plot**.

````python
kl_to = dt.loc['KERALA',['name_of_city', '0-6_population_total', 'literates_total', 'total_graduates','population_total']]
kl_to.set_index('name_of_city', inplace= True)
#kl_to = kl_to.transpose()
````

Area plots are stacked by default. And to produce a stacked area plot, each column must be either all positive or all negative values (any NaN values will defaulted to 0). To produce an unstacked plot, pass `stacked=False`. 

````python
kl_to.plot(kind='area', 
             stacked=False,
             figsize=(20, 10), # pass a tuple (x, y) size
             )

plt.title('State-wise Population')
plt.ylabel('Population')
plt.xlabel('District Name')

plt.show()
````

The unstacked plot has a default transparency (alpha value) at 0.5. We can modify this value by passing in the `alpha` parameter.

````python
kl_to.plot(kind='area', 
             alpha=0.75, # 0-1, default value a= 0.5
             stacked=True,
             figsize=(20, 10), # pass a tuple (x, y) size
             )

plt.title('State-wise Population')
plt.ylabel('Population')
plt.xlabel('District Name')

plt.show()
````

Let us see the same with another state.

````python
mh_to = dt.loc['MAHARASHTRA',['name_of_city','population_total', '0-6_population_total', 'literates_total', 'total_graduates']]
mh_to.set_index('name_of_city', inplace= True)
#mh_to = mh_to.transpose()
mh_to.plot(kind='area', 
             stacked=False,
             figsize=(10, 5), # pass a tuple (x, y) size
             )

plt.title('State-wise Population')
plt.ylabel('Population')
plt.xlabel('District Name')

plt.show()
````

### Two types of plotting

As we discussed in the video lectures, there are two styles/options of ploting with `matplotlib`. Plotting using the Artist layer and plotting using the scripting layer.

**Option 1: Scripting layer (procedural method) - using matplotlib.pyplot as 'plt' **

You can use `plt` i.e. `matplotlib.pyplot` and add more elements by calling different methods procedurally; for example, `plt.title(...)` to add title or `plt.xlabel(...)` to add label to the x-axis.
```python
    # Option 1: This is what we have been using so far
      plt.title('State-wise Population')
      plt.ylabel('Population')
      plt.xlabel('District Name')
```

**Option 2: Artist layer (Object oriented method) - using an `Axes` instance from Matplotlib (preferred) **

You can use an `Axes` instance of your current plot and store it in a variable (eg. `ax`). You can add more elements by calling methods with a little change in syntax (by adding "*set_*" to the previous methods). For example, use `ax.set_title()` instead of `plt.title()` to add title,  or `ax.set_xlabel()` instead of `plt.xlabel()` to add label to the x-axis. 

This option sometimes is more transparent and flexible to use for advanced plots (in particular when having multiple plots.

````python
# option 2: preferred option with more flexibility
ax = kl_to.plot(kind='area', alpha=0.35, figsize=(20, 10))

ax.set_title('State-wise Population')
ax.set_ylabel('Population')
ax.set_xlabel('District Name')
````

Let us use the scripting layer to create a stacked area plot of the 5 top states for effective_literacy_rate_total and plot the following(literates_male', 'literates_female', 'male_graduates', 'female_graduates'). Use a transparency value of 0.45.


Let us sum up the values in each district to their respective State to get our required plot.

````python
lis = dt.groupby('state_name').sum() # grouping to get the data by State-wise
lis.sort_values(by = 'effective_literacy_rate_total', ascending=False, axis=0, inplace=True) #sorting to get the top values
lis = lis.head(5)
````

````python
lis = lis.loc[:,['literates_male', 'literates_female', 'male_graduates', 'female_graduates']]
#lis = lis.transpose()
lis.plot(kind='area', 
             alpha=0.45, # 0-1, default value a= 0.5
             stacked=False,
             figsize=(20, 10),
            )

plt.title('5 Top states for effective_literacy_rate')
plt.ylabel('Population')
plt.xlabel('States')

plt.show()
````
