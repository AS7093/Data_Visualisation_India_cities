# Histograms<a id="8"></a>
A histogram is a way of representing the *frequency* distribution of numeric dataset. The way it works is it partitions the x-axis into *bins*, assigns each data point in our dataset to a bin, and then counts the number of data points that have been assigned to each bin. So the y-axis is the frequency or the number of data points in each bin. Note that we can change the bin size and usually one needs to tweak it so that the distribution is displayed nicely.

Import `Matplotlib`

```python
# use the inline backend to generate the plots within the browser
%matplotlib inline 

import matplotlib as mpl
import matplotlib.pyplot as plt

mpl.style.use('ggplot') # optional: for ggplot-like style

# check for latest version of Matplotlib
print ('Matplotlib version: ', mpl.__version__) # >= 2.0.0
```

Before we proceed with creating the histogram plot, let's first examine the data split into intervals. To do this, we will us **Numpy**'s `histrogram` method to get the bin ranges and frequency counts

```python
lis = dt.groupby('state_name').sum() # grouping to get the data by State-wise
lis.sort_values(by = 'effective_literacy_rate_total', ascending=False, axis=0, inplace=True) #sorting to get the top values
lis.head()
```

```python
#lis['population_total'] = lis['population_total']/1000
lis['population_total']
```

```python
# np.histogram returns 2 values
count, bin_edges = np.histogram(lis['population_total'])

print(count) # frequency count
print(bin_edges) # bin ranges, default = 10 bins
```

By default, the `histrogram` method breaks up the dataset into 10 bins. The figure below summarizes the bin ranges and the frequency distribution of population in states. We can see that:
* 15 State contributed between 0 to 100608 people 
* 11 countries contributed between 100608. to 3862760.8 people
* 1 country contributed between 3862760.8  to 7624913.6 people, and so on..

We can easily graph this distribution by passing `kind=hist` to `plot()`.

```python

lis['population_total'].plot(kind='hist', figsize=(10, 7))

plt.title('Histogram of Population in Different States from the sensus') # add a title to the histogram
plt.ylabel('Number of States') # add y-label
plt.xlabel('Number of people') # add x-label

plt.show()
```

In the above plot, the x-axis represents the population range of people intervals of 10068. The y-axis represents the number of States that contributed to the aforementioned population. 

Notice that the x-axis labels do not match with the bin size. This can be fixed by passing in a `xticks` keyword that contains the list of the bin sizes, as follows:

```python

lis['population_total'].plot(kind='hist', figsize=(8, 5), xticks=bin_edges)

plt.title('Histogram of Population in Different States from the sensus') # add a title to the histogram
plt.ylabel('Number of States') # add y-label
plt.xlabel('Number of people') # add x-label

plt.show()
```

# Bar Charts (Dataframe) <a id="10"></a>

A bar plot is a way of representing data where the *length* of the bars represents the magnitude/size of the feature/variable. Bar graphs usually represent numerical and categorical variables grouped in intervals. 

To create a bar plot, we can pass one of two arguments via `kind` parameter in `plot()`:

* `kind=bar` creates a *vertical* bar plot
* `kind=barh` creates a *horizontal* bar plot

**Vertical bar plot**

In vertical bar graphs, the x-axis is used for labelling, and the length of bars on the y-axis corresponds to the magnitude of the variable being measured. Vertical bar graphs are particuarly useful in analyzing time series data. One disadvantage is that they lack space for text labelling at the foot of each bar. 

Let's compare the number the literates of each States.

```python
lis = dt['literates_total']
lis = lis.groupby('state_name').sum()
lis
```

```python
# step 2: plot data
lis.plot(kind='bar', figsize=(20, 10))

plt.xlabel('States') # add to x-label to the plot
plt.ylabel('Population') # add y-label to the plot
plt.title('State-wise Literates') # add title to the plot

plt.show()
```

```python
lis = dt[['literates_total', 'literates_male', 'literates_female']]
lis = lis.groupby('state_name').sum()
lis = lis.head()
lis
```

```python
#create the plot
lfm = lis.plot(kind='bar', 
            figsize=(20, 8), 
            width=0.8, 
            color=['#5cb85c', '#5bc0de', '#d9534f'],
           )

lfm.set_title("State-wise Literates with male and female", fontsize=16)
lfm.spines['left'].set_visible(False)
lfm.spines['top'].set_visible(False)
lfm.spines['right'].set_visible(False)
lfm.axes.get_yaxis().set_visible(False)
lfm.tick_params(labelsize=14)
lfm.legend(fontsize=14)
for p in lfm.patches:
    lfm.annotate(np.round(p.get_height(),decimals=2), 
                (p.get_x()+p.get_width()/2., p.get_height()), 
                ha='center', 
                va='center', 
                xytext=(0, 10), 
                textcoords='offset points',
                fontsize = 14
               )
plt.show()
```
